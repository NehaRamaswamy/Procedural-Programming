### Write a program to print the outputs of 'FIZZ' , 'BUZZ' and 'FIZZBUZZ' per respective numbers.

```python
for val in range(1,100):
    if val % 3 == 0 and val % 5 == 0:
        print(f'{val} FIZZBUZZ')
    elif val % 3 == 0:
        print(f'{val} FIZZ')
    elif val % 5 == 0:
        print(f'{val} BUZZ')
    else:
        print(val)
```
